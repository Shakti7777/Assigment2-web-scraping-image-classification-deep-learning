```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

html = urlopen('https://www.amazon.in/b/ref=s9_acss_bw_cg_WWDTIMPC_2a1_w?node=17186599031&pf_rd_m=A1K21FY43GMZF8&pf_rd_s=merchandised-search-5&pf_rd_r=1C91CK7XYZXAX26P40VY&pf_rd_t=101&pf_rd_p=79601a8b-7d7e-4e08-84fe-e82957496171&pf_rd_i=1968256031')
bs = BeautifulSoup(html, 'html.parser')
images = bs.find_all('img', {'src':re.compile('.jpg')})
for image in images: 
    print(image['src']+'\n')
```
